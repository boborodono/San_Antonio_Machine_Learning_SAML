```python
# Imprting dependencies
import pandas as pd
from sklearn.preprocessing import OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
```

```python
medical_df = pd.read_csv('../Resources/Cleaned_Dataset/clean_stroke.csv')
print(medical_df.shape)
medical_df.head(10)
```

```python
medical_df.isnull().sum()
```

```python
medical_df.dtypes
```

```python
object_columns = medical_df.dtypes[medical_df.dtypes == 'object'].index.tolist()
object_columns
```

```python
medical_df[object_columns].nunique()
```

```python
medical_df['Work_Type'].value_counts()
```

```python
# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit & transform OneHotEncoder using categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(medical_df[object_columns]))

# Add encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(object_columns)
print(encode_df.shape)
encode_df.head()
```

```python
medical_df = medical_df.merge(encode_df, left_index=True, right_index=True).drop(object_columns, 1)
medical_df.head(10)
```

```python
medical_df.isnull().sum()
```

```python
medical_df.dtypes
```

```python
# Split preprocessed data into features & target arrays
y = medical_df['Stroke'].values
X = medical_df.drop(['Stroke'], 1).values

# Split into training & testing data sets
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=2, stratify=y)
```

```python
# Create a random forest classifier model
rf_model = RandomForestClassifier(n_estimators=128, random_state=2)

# Fit the random forest classifier model
rf_model = rf_model.fit(X_train, y_train)

# Evaluate the random forest classifier model
y_pred = rf_model.predict(X_test)
print(f' Random forest predictive accuracy: {accuracy_score(y_test, y_pred):.3f}')
```
